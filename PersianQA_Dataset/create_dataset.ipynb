```python
import json
import pandas as pd
```

```python
def process_json(file_path):
    contexts_data = []
    questions_data = []

    with open(file_path, "r", encoding="utf-8") as file:
        json_data = json.load(file)
        for article in json_data["data"]:
            for paragraph in article["paragraphs"]:
                context = paragraph["context"]
                context_id = hash(context) 
                
                contexts_data.append({"context_id": context_id, "context": context})
                
                for qas in paragraph["qas"]:
                    if not qas["is_impossible"]:
                        questions_data.append({"context_id": context_id, "question": qas["question"]})
    
    return contexts_data, questions_data
```

```python
contexts_train, questions_train = process_json("E:/Financial_Chatbot/PersianQA_Dataset/pqa_train.json")
contexts_train2, questions_train2 = process_json("E:/Financial_Chatbot/PersianQA_Dataset/train.json")

contexts_test, questions_test = process_json("E:/Financial_Chatbot/PersianQA_Dataset/pqa_test.json")
contexts_test2, questions_test2 = process_json("E:/Financial_Chatbot/PersianQA_Dataset/test.json")

contexts_combined = contexts_train + contexts_test + contexts_train2 + contexts_test2
questions_combined = questions_train + questions_test + questions_train2 + questions_test2

contexts_df = pd.DataFrame(contexts_combined)
questions_df = pd.DataFrame(questions_combined)
```

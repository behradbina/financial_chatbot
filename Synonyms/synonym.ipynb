```python
import pandas as pd
```

```python
df = pd.read_excel('data/Original_data/VirastaranMotaradefMotazad.xlsx', header=None, skiprows=2)
df.drop(columns=[df.columns[1]], inplace=True)

df = df.fillna('')

df = df.map(lambda cell: cell.split("≠")[0].strip())
df = df.map(lambda cell: cell.replace("،", ","))
df = df.map(lambda cell: cell.replace("\u200c", " "))
```

```python
df['synonyms'] = df.apply(lambda row: ', '.join(row[row != '']) if not row.isnull().all() else '', axis=1)
```

```python
synonyms_list = df['synonyms'].to_list()
synonyms_list
```

```python
df['synonym_count'] = df['synonyms'].apply(lambda x: len(x.split(', ')) if x else 0)

df['synonym_count'].max(), df['synonym_count'].min()
```

```python
len(df), len(df[df['synonym_count']==1]), len(df[df['synonym_count'] > 1])
```

```python
df = df[df['synonym_count'] > 1]
```

```python
len(df), len(df[df['synonym_count']==1]), len(df[df['synonym_count'] > 1])
```

```python
df[['synonyms']].to_csv('data/synonyms.csv', index=False, header=True, encoding='utf-8-sig')
```

```python
with open('data/synonyms.txt', 'w', encoding='utf-8-sig') as file:
    for item in synonyms_list:
        file.write("%s\n" % item)
```

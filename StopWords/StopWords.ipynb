```python
import os
import glob
import json
```

```python
directory = "E:/Financial_Chatbot/StopWords"

file_paths = glob.glob(os.path.join(directory, "*.txt"))

unique_words = set()

for file_path in file_paths:
    with open(file_path, 'r') as file:
        print(file_path)
        unique_words.update(line.strip() for line in file.readlines())

unique_word_list = sorted(unique_words)
```

```python
with open("Persian_Stop_Words.txt", "w", encoding="utf-8-sig") as file:
    for word in unique_word_list:
        file.write(word + "\n")  # Write each word on a new line

print("Stop words saved to stopwords.txt with UTF-8 encoding")
```
